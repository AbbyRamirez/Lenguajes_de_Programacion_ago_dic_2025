```python
import matplotlib.pyplot as plt


archivo = 'P2c5.txt'
x = []
y = []
with open(archivo, 'r') as f:
    for linea in f:
        partes = linea.strip().replace(',', ' ').split()
        if len(partes) == 2:
            try:
                xi = float(partes[0])
                yi = float(partes[1])
                x.append(xi)
                y.append(yi)
            except ValueError:
                print(f"Valores no válidos en la línea: {linea.strip()}")
plt.plot(x,y, marker='o', linestyle='-', color='grey')
plt.title('Angulo de 10° y vo de 2m/s a partir de (15,10) ')
plt.xlabel('alcance')
plt.ylabel('altura')
plt.grid(True)
plt.xlim(0,20)
plt.ylim(0,12)
plt.show()
```
