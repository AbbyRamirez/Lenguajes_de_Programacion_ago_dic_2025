```python
import matplotlib.pyplot as plt

# Ruta al archivo de texto
archivo = 'Pc5.txt'

# Listas para almacenar los valores x e y
x = []
y = []

# Leer el archivo
with open(archivo, 'r') as f:
    for linea in f:
        # Quitar espacios y dividir por espacio, coma o tab
        partes = linea.strip().replace(',', ' ').split()
        if len(partes) == 2:
            try:
                xi = float(partes[0])
                yi = float(partes[1])
                x.append(xi)
                y.append(yi)
            except ValueError:
                print(f"Valores no válidos en la línea: {linea.strip()}")

# Graficar
plt.plot(x,y, marker='o', linestyle='-', color='red')
plt.title('Angulo de 70° y vo de 15m/s')
plt.xlabel('alcance')
plt.ylabel('altura')
plt.grid(True)
plt.xlim(1, 5)  # Establece el rango del eje X
plt.ylim(2.5, 5)   # Establece el rango del eje Y
plt.show()
```
